```python
def geo_location(address):
    # get geo location of address
    geolocator = Nominatim(user_agent="TX_Rest")
    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude
    return latitude,longitude 
```

```python
def get_venues(lat,lng):
    
    #set variables
    radius=100000
    LIMIT=1000
    CLIENT_ID = 'JOR34Q4RQEBF5VFEMPY1LZ2CJET2U1IUVX3SE21CVINHDHIS' # your Foursquare ID
    CLIENT_SECRET = '5RYIMZB4EKE01XQRQWRKPNLEL3FYINQNBXSX3AU0IXUK4Q3E' # your Foursquare Secret
    VERSION = '20191001' # Foursquare API version
    search_query = 'indian restaurant'
    
    #url to fetch data from foursquare api
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&query={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT,
            search_query)
    #print(url)
    # get all the data
    results = requests.get(url).json()
    #print(results)
    venue_data=results["response"]['groups'][0]['items']
    venue_details=[]
    for row in venue_data:
        try:
            venue_id=row['venue']['id']
            venue_name=row['venue']['name']
            venue_category=row['venue']['categories'][0]['name']
            venue_zip=row['venue']['location']['postalCode']
            venue_lat=row['venue']['location']['lat']
            venue_long=row['venue']['location']['lng']
            if venue_category.lower().find("restaurant") != -1:
                venue_details.append([venue_id,venue_name,venue_category,venue_zip,venue_lat,venue_long])
        except KeyError:
            pass
        
    column_names=['ID','Name','Category','Zip','Latitude','Longitude']
    df = pd.DataFrame(venue_details,columns=column_names)
    return df
```

```python
loc=geo_location('Austin')
get_venues(loc[0], loc[1])
```

```python
map_austin = folium.Map(loc, zoom_start=11)
dataframe=get_venues(loc[0], loc[1])
# add markers to map
for lat, lng, postalcode, name in zip(dataframe['Latitude'],dataframe['Longitude'],dataframe['Zip'],dataframe['Name']):
    label = '{}, {}'.format(name, postalcode)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_austin)  
    #print(label)
map_austin
```
